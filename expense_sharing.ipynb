```python
import numpy as np
from prettytable import PrettyTable

friends=[]
expense_matrix=None
history=[]

#entering friends name
no_of_friends=int(input("Enter the count :"))
for i in range(no_of_friends):
    friends.append(input(f"Enter the name of friend {i+1} :"))
print(friends) 

#expense matrix
expense_matrix=np.zeros((no_of_friends,no_of_friends))

#splitting expenses
def expense():
    payerName=input("Enter the name of payer :")
    amountPayed=float(input("Enter the amount payed :"))
    beneficiaries=input("Enter the name of beneficiaries by comma seperated :").split(",")
    beneficiaries=[x.strip() for x in beneficiaries]
    
    if payerName not in friends:
        print("Friend not found!")
        return
    
    payerId=friends.index(payerName)
    share=amountPayed/len(beneficiaries)
    
    #splitting amount to all friends who are beneficiaries
    for person in beneficiaries:
        personId=friends.index(person)
        expense_matrix[payerId][personId] +=share
    
    #history update
    history.append({"payer": payerName,
                    "amount": amountPayed,
                    "beneficiaries": beneficiaries
                   })
    print("Expense added successfully!\n")
    
#showing history
def showhistory():
    if not history:
        print("No history found!!\n")
        return
    print("Transaction history!!\n")
    for i,item in enumerate(history,start=1):
        print(f"Expense {i} :")
        print("Payer:", item["payer"])
        print("Amount:", item["amount"])
        print("Details:", item["beneficiaries"])
        print("-" * 40)

#calculation of the amount of each friend
def calculate_settlements():
    total_paid=np.sum(expense_matrix,axis=1)
    total_owned=np.sum(expense_matrix,axis=0)
    
    net_balances=total_paid-total_owned
    return net_balances

#print the settlement
def display_settlements():
    settlements = calculate_settlements()
    table = PrettyTable()
    table.field_names = ["Friends","Settlements"]
    
    for i, friend in enumerate(friends):
        if settlements[i]>0:
            table.add_row([friend,f"should recive rupees :{settlements[i]:.2f}"])
        
        elif settlements[i]<0:
            table.add_row([friend,f"Owns rupees :{-settlements[i]:.2f}"])
        
        else:
            table.add_row([friend,"Settled"])
    
    print("\nFinal settlements\n")
    print(table)
    

#suggest settlements
def suggest_settlements():
    settlements=calculate_settlements()
    
    creditors=[
        (friends[i],amt)
        for i,amt in enumerate(settlements)
        if amt>0
    ]
    
    debitors=[
        (friends[i],-amt)
        for i,amt in enumerate(settlements)
        if amt<0
    ]
    
    transactions = []
    
    while debitors and creditors:
        debitor,deb_amt=debitors.pop(0)
        creditor,cred_amt=creditors.pop(0)
        
        payments=min(deb_amt,cred_amt)
        
        transactions.append((debitor,creditor,payments))
        
        deb_amt -= payments
        cred_amt -= payments
        
        if deb_amt>0:
            debitors.insert(0,(debitor,deb_amt))
        
        if cred_amt>0:
            creditors.insert(0,(creditor,cred_amt))
        
    print("\n Transaction Successfull\n")
        
    if not transactions:
        print("Everyone is settled.")
        
    else:
        for debitor,creditor, amount in transactions:
            print(f"{debitor} should pay ₹{amount:.2f} to {creditor}")

#show expense matrix
def show_expense():
    print("\n Expense Matrix !!")
    table=PrettyTable()
    table.field_names=["Paid By"]+friends
    
    for i,person in enumerate(friends):
        row=[person]
        row.extend([f"{x:.2f}" for x in expense_matrix[i]])
        table.add_row(row)
    
    print(table)
    
#Menu
while True:

    print("\n===== GOOGLE PAY EXPENSE SHARING =====")

    print("1. Add Expense")
    print("2. Show Transaction History")
    print("3. Show Settlements")
    print("4. Suggest Payments")
    print("5. Show Expense Matrix")
    print("6. Exit")

    choice = input("\nEnter choice: ")

    if choice == '1':
        expense()

    elif choice == '2':
        showhistory()

    elif choice == '3':
        display_settlements()

    elif choice == '4':
        suggest_settlements()

    elif choice == '5':
        show_expense()

    elif choice == '6':
        print("Thank You!")
        break

    else:
        print("Invalid Choice")
```
